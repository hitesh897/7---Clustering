### Hierarchial Clustering

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crime = pd.read_csv("crime_data.csv")
crime
```

```python
#Normalize Data Function
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
crime_norm=norm_func(crime.iloc[:,1:])
crime_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(crime_norm,method='average'))
```

```python
# Create Clusters
h_clusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
h_clusters
```

```python
y_hc=h_clusters.fit_predict(crime_norm)
y_hc
```

```python
crime['clusterid']=h_clusters.labels_
crime
```

##  KMeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime1=pd.read_csv("crime_data.csv")
crime1
```

```python
# Normalized Data Function
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
crime1_norm=norm_func(crime1.iloc[:,1:])
```

```python
# Elbow Curve
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(crime1_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
    
```

```python
# Select 4 clusters from Elbow Curve
model=KMeans(n_clusters=4)
model.fit(crime1_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['clusterid']=x
crime1
```

```python
crime1.groupby('clusterid').agg(['mean'])
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2=pd.read_csv('crime_data.csv')
crime2
```

```python
crime2.info()
```

```python
crime3=crime2.iloc[:,1:5]
crime3
```

```python
crime3.values
```

```python
# Standardizing the values
stscaler=StandardScaler().fit(crime3.values)
x=stscaler.transform(crime3.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2, min_samples=5)
dbscan.fit(x)
```

```python
# Noisy sample will be represented by -1
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['clusters'])
```

```python
cl
```

```python
pd.concat([crime2,cl],axis=1)
```
